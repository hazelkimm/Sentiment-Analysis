#### TF-IDF
- pick highly positive review / highly negative review respectively
- find out how important words are in the sentence

```python
# randomly pick one review that is highly positive (score over 0.95)
pos = names.loc[names['compound'] > 0.95].sample()
# make review list as string
bagpos = ' '.join(str(e) for e in pos["text_token"])
bagpos
```

```python
# randomly pick one review that is highly negative (score below -0.95)
neg = names.loc[names['compound'] < -0.95].sample()
# make review list as string
bagneg = ' '.join(str(e) for e in neg["text_token"])
bagneg
```

```python
#remove duplicate words
uniqueWords = set(bagpos).union(set(bagneg))
```

```python
#create a dictionary of words and their occurrence for each document in the corpus(collection of documents)
numpos = dict.fromkeys(uniqueWords, 0)
for word in bagpos:
    numpos[word] += 1
numneg = dict.fromkeys(uniqueWords, 0)
for word in bagneg:
    numneg[word] += 1
bagpos
```

```python
#implement term frequency in python
def computeTF(wordDict, bagOfWords):
    tfDict = {}
    bagOfWordsCount = len(bagOfWords)
    for word, count in wordDict.items():
        tfDict[word] = count / float(bagOfWordsCount)
    return tfDict
```

```python
tfpos = computeTF(numpos, bagpos)
tfneg = computeTF(numneg, bagneg)
```

```python
#implement inverse data frequency
def computeIDF(documents):
    import math
    N = len(documents)
    
    idfDict = dict.fromkeys(documents[0].keys(), 0)
    for document in documents:
        for word, val in document.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log(N / float(val))
    return idfDict
```

```python
idfs = computeIDF([numpos, numneg])
```

```python
#implement TF-IDF by multiplying TF with IDF
def computeTFIDF(tfBagOfWords, idfs):
    tfidf = {}
    for word, val in tfBagOfWords.items():
        tfidf[word] = val * idfs[word]
    return tfidf
```

```python
#compute TF-IDF scores for all the words in the corpus
tfidfpos = computeTFIDF(tfpos, idfs)
tfidfneg = computeTFIDF(tfneg, idfs)
score = pd.DataFrame([tfidfpos, tfidfneg])
score
```
